```python
import pandas as pd
df = pd.read_excel('/content/Север на проверку.xls')
```

```python
lst = []

for i in df['Фильтр']:
  if pd.isnull(i) == False:
    lst.append(i)
  else:
    break
```

```python
listWrongs = []

for i in df['Офис, ответственный за доставку']:
  if i in lst:
    continue
  else:
    listWrongs.append(i)

listWrongsU = list(set(listWrongs))
listWrongsUClear = [x for x in listWrongsU if str(x) != 'nan']
```

```python
rslt_df = (df.loc[df['Офис, ответственный за доставку'].isin(listWrongsUClear)])
```

```python
print('\nResult dataframe :\n',
      rslt_df)
```

```python
rslt_df.to_excel('Засылы СЦ постобработка.xlsx', index=False)
```
